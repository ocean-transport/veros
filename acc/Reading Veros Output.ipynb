```python
import xarray as xr

ds = xr.open_dataset("acc.snapshot.nc", engine="h5netcdf")

# plot surface velocity at the last time step included in the file
u_surface = ds.u.isel(Time=-1, zt=-1)
u_surface.plot.contourf()
```
